### Píxel color

En este caso el objeto a detectar son regiones de píxeles conectadas que tienen un rango de color específicos. Por ejemplo: la piel.

<img src="./images/manos.png" alt="manos">

<img src="./images/pixelcolor1.png" alt="pixelcolor1">

¿Qué pasa si tenemos cambios en la intensidad de la luz?

<img src="./images/pixelcolorintensity.png" alt="intensity">

Podemos corregirlo pasando a un espacio de color invariante a la intensidad:

$$(R, G, B)= ( \frac{R}{R+G+B},\frac{G}{R+G+B},\frac{B}{R+G+B} ) $$

<img src="./images/pixelcolorintensity2.png" alt="intensity2">

Muchas veces además las imágenes presentan cambios de color producto de cambios en la iluminación o en la cámara:

Esto se puede solucionar normalizando a cada canal por su valor máximo. A este método se lo llama white patching.

$$(R, G, B)= ( \frac{255.R}{R_{max}(I)},\frac{255.G}{G_{max}(I)},\frac{255.B}{B_{max}(I)} ) $$

<img src="./images/pixelcolorcolor.png" alt="colorproblem">

Para la generación de ventanas de salida se buscarán regiones conectadas de píxel de clase piel. 

#### Generación de ventanas:

Algoritmo de etiquetaje: Dada una imagen binaria retorna un imagen con etiquetas numéricas, de manera que todos los píxeles que pertenecen a una misma región conectada compartan la misma etiqueta.

Dos recorridos de la imagen (Un recorrido pasa por cada píxel de la imagen de izquierda a derecha y de arriba abajo):

1) Etiquetar cada píxel 1 de la imagen atendiendo a sus vecinos superior e izquierdo según la conectividad (4-conexo u 8-conexo) y guardar posibles equivalencias.

* Caso 1 (No tiene vecinos): Se asigna una etiqueta nueva.
* Caso 2 (Tiene solo un vecino): Se asigna la etiqueta del vecino.
* Caso 3 (Tiene mas de un vecino): Se asigna la etiqueta de alguno de los vecinos y se guardan las equivalencias.

2) Resuelve todas las equivalencias detectadas de etiquetas y selecciona una etiqueta para cada equivalencia, el segundo recorrido las resuelve y asigna la etiqueta seleccionada.

<img src="./images/ejemploetiquetado.png" alt="ejemploetiquetado">

### Template Matching (basado en la diferencia)

* Cálculo de la diferencia: Para cada punto de la imagen se asigna un valor que representa la semejanza entre el patrón y la ventana de la imagen, la cual tienen el mismo tamaño que el patrón.
* Extracción de mínimos: El punto de la imagen que presente mínimo valor de las diferencias, nos dará la localización del objeto encontrado en la imagen. Se pueden detectar varias apariciones de la imagen buscando varios mínimos locales.

<img src="./images/TM-diff1.png" alt="diff1">

<img src="./images/TM-diff2.png" alt="diff2">

Esta forma de detectar templates tiene el problema de que es sensible a variaciones de intensidad. Otra opción cuyo comportamiento es invariante a la intensidad es el template matching basado en correlación:

<img src="./images/TM-corr1.png" alt="corr1">

<img src="./images/TM-corr2.png" alt="corr2">


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('messi5.jpg',0)
img2 = img.copy()
template = cv2.imread('template.jpg',0)
w, h = template.shape[::-1]

# All the 6 methods for comparison in a list
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for meth in methods:
    img = img2.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)

    cv2.rectangle(img,top_left, bottom_right, 255, 2)
    plt.figure(figsize=(16,4))
    plt.subplot(121),plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.suptitle(meth)

    plt.show()
```

### Local Binary Pattern (LBP) [paper](http://www.outex.oulu.fi/publications/pami_02_opm.pdf) [doc](https://www.pyimagesearch.com/2015/12/07/local-binary-patterns-with-python-opencv/)

<img src="./images/lbp1.png" alt="lbp1">

Al implementar este procedimiento tenemos un descriptor invariante al cambio de intensidad.

OpenCV no implementa lbp de forma abierta, si bien lo usa para algunas funciones de face recognition. LBP viene incluído en scikit-image:

```python
from skimage import feature
import numpy as np
image = cv2.imread("messi5.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
lbp=feature.local_binary_pattern(gray,8,1,"default")
plt.figure(figsize=(16,4))
plt.subplot(121),plt.imshow(gray,cmap = 'gray')
plt.title('Orginal'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(lbp,cmap = 'gray')
plt.title('Procesada'), plt.xticks([]), plt.yticks([])
plt.suptitle("LBP Básico")
plt.show()
```

Una característica de este descriptor es que la representación de los datos ocupa lo mismo que los datos originales. Una variante que se utiliza es la conocida como LBP-U (Uniforme).

### LBP Uniforme

Como primera medida, consiste en calcular con la medida de uniformidad de cada uno de los bytes LBP. Se define a la medida de uniformidad como la cantidad de veces que se presentan transiciones de bit en el patrón LBP. Para el caso anterior: 0b1111111111000111, la medida de uniformidad es 2, ya que presenta dos transiciones.

* La cantidad de patrones de uniformidad 0 son dos: 0b00000000 y 0b11111111
* La medida de uniformidad es siempre un número par

Luego de calcular las medidas de uniformidad:

* A los patrones cuya medida de uniformidad U=0 o U=2 les reasignamos un código de patrón individual.
* A los patronos cuya medida de uniformidad es mayor a 2 se le reasigna el mismo código (pasan a ser indistiguibles)

LBP: 256 patrones
LBP-U: 58+1 patrones


```python
lbpu=feature.local_binary_pattern(gray,8,1,"nri_uniform")
plt.figure(figsize=(16,4))
plt.subplot(121),plt.imshow(gray,cmap = 'gray')
plt.title('Orginal'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(lbpu,cmap = 'gray')
plt.title('Procesada'), plt.xticks([]), plt.yticks([])
plt.suptitle("LBP Uniforme")
plt.show()
```

Por ejemplo, para una ventana de detección de 128 x 64, necesito almacenar 8192 descriptores. En cambio, en el caso de que utilice el histograma LBP-U, solo necesitaré guardar 59.  
Esta técnica además tiene como ventaja que el alineamiento de la ventana y el template no tiene por qué se preciso.

## Generación de candidatos

Hemos dicho que contamos con un generador de candidatos, y un clasificador de candidatos. Para terminar de implementar el generador de candidatos, debemos definir una ventana de tamaño predefinido (lo llamaremos canónico) y la iremos moviendo con pasos $\Delta X$ en el eje horizontal y pasos $\Delta Y$ en el eje vertical. 
Iremos recortando la imagen original en sucesivas ventanas de tamaño canónico y con ello generaremos nuestros candidatos que recibirá el clasificador. A este procesimiento se le llama: Sliding Window.

<img src="./images/sliding-window.png" alt="histograma lbp-u por bloques">

Hay que tener en cuenta que si el tamaño del paso es menor que el tamaño de la ventana, se pueden tomar algunas medidas para reducir el cálculo, por ejemplo, primero calcular los códigos LBP para toda la imagen, y luego ir haciendo los histogramas para cada paso del sliding window.

El problema que se tiene en este punto es que podemos querer detectar distintos tamaños del mismo objeto. A tal fin lo que se hace es procesar no solo la imagen original, sino la pirámide de imágenes gaussiana correspondiente a esa imagen. Se utiliza la pirámide gaussiana con el fin de evitar artefactos. 

<img src="./images/pyramid-sw.png" alt="piramyd sw">

Ahora lidiamos con el problema de múltiples detecciones. Teniendo en cuenta que se utilizan clasificadores cuya salida nos dan los niveles de confidencia (por ejemplo con un regresor logístico, SVM, etc), el problema se puede resolver con el método Non Maximum Supression:

1) Ordenamos todas las detecciones de mayor score a menor score y las guardamos en una lista S.  
2) Inicializamos una lista de detecciones G, vacía.  
3) Tomamos el primer elemento de S y nos fijamos si su ventana tiene mas de K overlap (por ej K=0.5) que alguna de las detecciones guardadas en G.  
4) Si no tiene mas de K overlap con un elemento de G, entonces hacemos una copia de la detección en G.  
5) Eliminamos la detección de S.  
6) Volvemos al punto 3 hasta que S quede vacío.  
7) Los elementos de G serán nuestras detecciones finales.
